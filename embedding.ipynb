```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent = [
    'I love programming in Python',
    'Python is great for data science',
    'I enjoy learning new programming languages',
    'Data science is fascinating and rewarding',
    'Python and R are popular in data science'
]
```

```python
vocab_size = 1000
```

```python
one_hot_rep = [one_hot(words, vocab_size) for words in sent]
one_hot_rep
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences
```

```python
sent_length = 8
embedded_docs = pad_sequences(one_hot_rep,maxlen=sent_length,padding='pre')
embedded_docs
```

```python
dim = 10
model= Sequential()
model.add(Embedding(vocab_size,dim,input_length=sent_length))
model.compile(optimizer='adam',loss='mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
